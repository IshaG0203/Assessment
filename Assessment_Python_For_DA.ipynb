```python
class Clinic_appointment:
  def __init__(self):
    self.time_slot = ["10am", "11am", "12pm", "2pm", "3pm"]
    self.doctors = ["Dr. Sharma", "Dr. Patel", "Dr. Mehta"]
    self.appointments=[]

  def book_appointment(self):
    patient_name = input("Enter patient name: ")
    patient_age = int(input("Enter patient age: "))
    patient_mobile_number = input("Enter patient Mobile Number: ")

   def show_time_slot(self):
        print("\nAvailable Time Slots:")
        for i, slot in enumerate(self.time_slot, start=1):
            print(f"{i}. {slot}")

    try:
        choice = int(input("Select time slot number: "))
        if not (1 <= choice <= len(self.time_slot)):
            print("Invalid slot selection. Please choose a number from the list.")
            return
        selected_slot = self.time_slot[choice - 1]
    except ValueError:
        print("Invalid input. Please enter a number.")
        return

    print("\nAvailable Doctors:")
    for index, doctor_val in enumerate(self.doctors):
      print(f'{index + 1}. {doctor_val}')

    try:
      doc_choice = int(input("Select doctor number: "))
      if not (1 <= doc_choice <= len(self.doctors)):
        print("Invalid doctor selection.")
        return
      selected_doctor = self.doctors[doc_choice - 1]
    except ValueError:
      print("Invalid input. Please enter a number.")
      return

    count = 0
    for A in self.appointments:
      if A["doctor"] == selected_doctor and A["slot"] == selected_slot:
        count += 1
    if count >= 3:
      print("Slot full for this doctor. Choose another slot.")
      return

    self.appointments.append({
        "name": patient_name,
        "age": patient_age,
        "mobile": patient_mobile_number,
        "doctor": selected_doctor,
        "slot": selected_slot
    })
    print("Appointment booked successfully!")

  def view_appointment(self):
    mobile = input("Enter mobile number: ")
    found = False
    for A in self.appointments:
      if A["mobile"] == mobile:
        print("\nðŸ“‹ Appointment Details")
        print(f"Name   : {A['name']}")
        print(f"Age    : {A['age']}")
        print(f"Doctor : {A['doctor']}")
        print(f"Time   : {A['slot']}")
        found = True
        break
    if not found:
      print("No appointment found.")

  def cancel_appointment(self):
    mobile = input("Enter mobile number: ")
    for A in self.appointments:
      if A["mobile"] == mobile:
        self.appointments.remove(A)
        print("Appointment cancelled successfully.")
        return
    print("No appointment found to cancel.")

clinic = Clinic_appointment()
def clinic_schedule():
  while True:
    print("\n--- Clinic Appointment System ---")
    print("1. Book Appointment")
    print("2. View Appointment")
    print("3. Cancel Appointment")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
      clinic.book_appointment()
    elif choice == "2":
      clinic.view_appointment()
    elif choice == "3":
      clinic.cancel_appointment()
    elif choice == "4":
      print("Thank you!")
      break
    else:
      print("Invalid choice")
```

```python
class Student_Admissions:
  def __init__(self):
    self.students = {}
    self.next_id_prefix = "E"
    self.next_id_number = 1001

  def new_admission(self):
    name = input("Enter student name: ")
    age = int(input("Enter student age(5-18): "))
    Class = int(input("Enter student Class(1-12): "))
    gardian_mobile_number = input("Enter gardian mobile number: ")

    if age < 5 or age > 18:
      print("Invalid Age, Age should be between 5 and 18.")
      return
    if Class < 1 or Class > 12:
      print("Invalid Class, Class should be between 1 and 12.")
      return
    if not len(gardian_mobile_number)==10:
      print("Invalid Mobile Number, Mobile number should be 10 digits.")
      return

    student_ID = f"{self.next_id_prefix}{self.next_id_number}"
    self.students[student_ID] = {
        "name": name,
        "age": age,
        "Class": Class,
        "gardian_mobile_number": gardian_mobile_number
    }
    self.next_id_number += 1
    print(f"Admission successful. Student ID: {student_ID}")

  def view_admission(self):
    student_id = input("Enter student ID: ")
    if student_id in self.students:
      print("\nðŸ“‹ Student Details")
      for key, value in self.students[student_id].items():
          print(f"{key.replace('_', ' ').title()} : {value}")
    else:
      print("Student not found.")

  def update_admission(self):
    student_id = input("Enter student ID: ")
    if student_id not in self.students:
      print("Student not found.")
      return
    print("1. Update Class")
    print("2. Update Guardian Mobile Number")
    choice = input("Enter choice: ")
    if choice == "1":
      new_Class = int(input("Enter new Class: "))
      if new_Class < 1 or new_Class > 12:
        print("Invalid Class, Class should be between 1 and 12.")
        return
      self.students[student_id]["Class"] = new_Class
      print(f"Student {student_id} Class updated to {new_Class}.")
    elif choice == "2":
      new_gardian_mobile_number = input("Enter new Guardian Mobile Number: ")
      if not len(new_gardian_mobile_number) == 10:
        print("Invalid Mobile Number, Mobile number should be 10 digits.")
        return
      self.students[student_id]["gardian_mobile_number"] = new_gardian_mobile_number
      print(f"Student {student_id} Guardian Mobile Number updated.")
    else:
        print("Invalid choice.")

  def remove_admission(self):
    student_id = input("Enter student ID to remove: ")
    if student_id not in self.students:
      print("Student not found.")
      return
    del self.students[student_id]
    print(f"Student {student_id} removed successfully.")

School = Student_Admissions()
def admission_schedule():
  while True:
    print("\n--- School Admission System ---")
    print("1. New Admission")
    print("2. View Admission")
    print("3. Update Admission")
    print("4. Remove Admission")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
      School.new_admission()
    elif choice == "2":
      School.view_admission()
    elif choice == "3":
      School.update_admission()
    elif choice == "4":
      School.remove_admission()
    elif choice == "5":
      print("Thank you!")
      break
    else:
      print("Invalid choice")
```

```python
class BusReservation:
    def __init__(self):
        self.routes = {
            "Mumbai to Pune": 500,
            "Delhi to Jaipur": 600,
            "Surat to Mumbai": 500,
            "Surat to Rajkot": 700
        }

        self.bookings = {}
        self.seats = {}
        self.ticket_no = 1001


        for route in self.routes:
            self.seats[route] = 0

    def show_routes(self):
        print("\nAvailable Routes:")
        i = 1
        for route, price in self.routes.items():
            print(f"{i}. {route} - Rs.{price}")
            i += 1

    def book_ticket(self):
        name = input("Enter passenger name: ")
        age = int(input("Enter age: "))
        mobile = input("Enter mobile number: ")

        self.show_routes()
        choice = int(input("Select route number: "))

        route_list = list(self.routes.items())

        if choice < 1 or choice > len(route_list):
            print("Invalid route selection!")
            return

        route, price = route_list[choice - 1]

        if self.seats[route] >= 40:
            print("Sorry, no seats available for this route.")
            return

        self.seats[route] += 1
        seat_no = self.seats[route]
        ticket_id = "T" + str(self.ticket_no)
        self.ticket_no += 1

        self.bookings[ticket_id] = {
            "Name": name,
            "Age": age,
            "Mobile": mobile,
            "Route": route,
            "Price": price,
            "Seat": seat_no
        }

        print("\nTicket Booked Successfully!")
        print("Ticket ID:", ticket_id)
        print("Seat Number:", seat_no)

    def view_ticket(self):
        ticket_id = input("Enter Ticket ID: ")

        if ticket_id in self.bookings:
            print("\nTicket Details:")
            for key, value in self.bookings[ticket_id].items():
                print(f"{key}: {value}")
        else:
            print("Ticket not found!")


    def cancel_ticket(self):
        ticket_id = input("Enter Ticket ID to cancel: ")

        if ticket_id in self.bookings:
            route = self.bookings[ticket_id]["Route"]
            self.seats[route] -= 1
            del self.bookings[ticket_id]
            print("Ticket cancelled successfully!")
        else:
            print("Invalid Ticket ID!")

    def menu(self):
        while True:
            print("\n--- Bus Reservation System ---")
            print("1. Show Available Routes")
            print("2. Book Ticket")
            print("3. View Ticket")
            print("4. Cancel Ticket")
            print("5. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.show_routes()
            elif choice == "2":
                self.book_ticket()
            elif choice == "3":
                self.view_ticket()
            elif choice == "4":
                self.cancel_ticket()
            elif choice == "5":
                print("Thank you! Exiting system.")
                break
            else:
                print("Invalid choice!")

bus = BusReservation()
bus.menu()
```
